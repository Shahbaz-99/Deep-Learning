Bag of Words in Natual Language Processing(NLP)

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'text': ['people watch campusx','campusx watch campusx','people write comment','campusx write comment'],'output':[1,1,0,0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df["text"])
```

```python
# Our Vocabs
cv.vocabulary_
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
```

```python
cv.fit_transform(["campusx watch and write comment"]).toarray()
```

N-Grams

N-Grams are the advanced version of Bag of Words

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'text':['people watch campusx','campusx watch campusx','people write comment','campusx write comment'],'output':[1,1,0,0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2)) # ngram range tell us how many merged word we need in a word!
```
