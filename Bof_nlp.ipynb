```python
import re

def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'',text)
```

Creating method to remove url from the datasets

```python
import re 
def remove_url(text):
    pattern = re.compile(r"https?://\S+|www\.\S+")
    return pattern.sub(r'',text)
```

```python
text1 = "check my notebook https://www.kaggle.com/compusx/notebook"
remove_url(text1)
```

```python
df['review'] = df['review'].apply(remove_url)
```

Creating methods to remove punctuations

```python
import string, time
string.punctuation
```

```python
exclude = string.punctuation
exclude
```

```python
def remove_punctuation(text):
    for char in text:
        if char in exclude:
            text = text.replace(char,'')
        
    return text    
```

Removed punctuations from the dataframe

```python
df['review'] = df['review'].apply(remove_punctuation)
```

```python
df['review'][2]
```
